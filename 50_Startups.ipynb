## 7. Model Validation

### 7.1 - Collinearity Check :

```python
#Calculate VIF (Varience Inflation Factor - to describe when two or more predictors in regression are highly correlated) : 
```

### VIF  = 1 - (1 - R-Square)  for all independent Variables

```python
rsq_rds = smf.ols('RDS ~ ADMS + MS',data = startups_data).fit().rsquared
vif_rds = 1/(1 -rsq_rds)

rsq_adms = smf.ols('ADMS ~ RDS + MS', data = startups_data).fit().rsquared
vif_adms = 1/(1 - rsq_adms)

rsq_ms = smf.ols('MS ~ RDS + ADMS', data = startups_data).fit().rsquared
vif_ms = 1/(1 - rsq_ms)

data = { 'Variables' : ['RDS','ADMS','MS'],
         'VIF'       : [vif_rds,vif_adms,vif_ms]
        
        }
Vif = pd.DataFrame(data)
Vif
```

```python
# There is no collinearity in this variables . 
```

## 8. Model Deletion


  

### Two techniques for model deletion :
      1. Cook's Distance - Regression Analysis to find influential outliers
      2. Leverage Value

#### 1 . Cook's Distance

```python
# If Cook's distance > 1, then it's an outlier

(c,_) = model.get_influence().cooks_distance
c
```

```python
# Plot the influencers using the stem plot
fig=plt.figure(figsize=(20,7))
plt.stem(np.arange(len(startups_data)),np.round(c,5))
plt.xlabel('Row Index')
plt.ylabel('Cooks Distance')
plt.show()
```

```python
# Index and value of influencer where C>0.5
np.argmax(c) , np.max(c)
```

#### 2 . Leverage Value 

```python
# Leverage Value using High Influence Points : Points beyond Leverage_cutoff value are influencers
influence_plot(model)
plt.show()
```

```python
# Leverage Cuttoff Value = 3*(k+1)/n ; k = no.of features/columns & n = no. of datapoints

k = startups_data.shape[1]
n = startups_data.shape[0]
leverage_cutoff = (3*(k+1))/n
leverage_cutoff
```

```python
startups_data[startups_data.index.isin([49])]
```

```python
## Improve the model 
```

```python
# Discard the data points which are influencers and reassign the row number (reset_index(drop=True))

startups_data = startups_data.drop(startups_data.index[[49]],axis=0).reset_index(drop=True)
startups_data
```

## Final Model

```python
while np.max(c)>0.5 :
    model = smf.ols("Profit ~ RDS + ADMS + MS", data = startups_data).fit()
    (c,_) = model.get_influence().cooks_distance
    c
    np.argmax(c) , np.max(c)
    startups_data = startups_data.drop(startups_data.index[[np.argmax(c)]],axis=0).reset_index(drop=True)
    startups_data
else:
    final_model = smf.ols("Profit ~ RDS + ADMS + MS", data = startups_data).fit()
    final_model.rsquared , final_model.aic
    print("Thus model accuracy is improved to",final_model.rsquared)
```
